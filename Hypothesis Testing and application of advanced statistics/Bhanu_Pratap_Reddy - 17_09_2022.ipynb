```python
#Declare all know Probabilities

p_leak_given_fire = 0.2
p_leak_given_mechanical = 0.5
p_leak_given_human_error = 0.1
p_leak_intersection_fire = 0.001
p_leak_intersection_mechanical = 0.0015
p_leak_intersection_human = 0.0012
```

```python
p_fire = round(p_leak_intersection_fire/p_leak_given_fire,4)

print(f"The Probability of a fire occurring is {p_fire} i.e. {p_fire*100}%")
```

```python
p_mechanical_failure = round(p_leak_intersection_mechanical/p_leak_given_mechanical,4)

print(f"The Probability of a mechanical failure occurring is {p_mechanical_failure} i.e. {p_mechanical_failure*100}%")
```

```python
p_human_error = round(p_leak_intersection_human/p_leak_given_human_error,4)

print(f"The Probability of human error happening is {p_human_error} i.e. {p_human_error*100}%")
```

```python
#Radiation leak can be given as

p_leak = round((p_fire*p_leak_given_fire)+(p_human_error*p_leak_given_human_error)+(p_mechanical_failure*p_leak_given_mechanical),4)

print(f"The Probability of Radiation leak occurring is {p_leak} i.e. {p_leak*100}%")
```

```python
#As per Bayes Theorem

p_fire_given_leak = round(((p_fire*p_leak_given_fire)/p_leak),4)

print(f"The Probability of fire occurring given leak has happened is {p_fire_given_leak} i.e. {p_fire_given_leak*100}%")
```

```python
#As per Bayes Theorem

p_mechanical_given_leak = round(((p_mechanical_failure*p_leak_given_mechanical)/p_leak),4)

print(f"The Probability of mechanical failure occurring given leak has happened is {p_mechanical_given_leak} i.e. {p_mechanical_given_leak*100}%")
```

```python
#As per Bayes Theorem

p_human_error_given_leak = round(((p_human_error*p_leak_given_human_error)/p_leak),4)

print(f"The Probability of human error occurring given leak has happened is {p_human_error_given_leak} i.e. {p_human_error_given_leak*100}%")
```

```python
#Declare given variables of normal distribution

mean_breaking_str = 5
std_dev_dist = 1.5
```

```python
#Prepare a dataset for visal representation
breaking_strength_dist= np.random.normal(loc=mean_breaking_str, scale=std_dev_dist, size=500) 
breaking_strength_dist_min = breaking_strength_dist.min()
breaking_strength_dist_max =breaking_strength_dist.max()

#x-axis ranges from min to max with 500 steps
x = np.arange(breaking_strength_dist_min,breaking_strength_dist_max , (breaking_strength_dist_max-breaking_strength_dist_min)/500 )
br_str_dist_pdf = norm.pdf(x, mean_breaking_str, std_dev_dist)

#plotting the distribution
plt.figure(figsize=(15,8))
plt.plot(x, br_str_dist_pdf)
# plt.grid()
plt.title("Gunny Bag Breaking Strength Distribution",fontsize=15, weight="bold")

plt.xlabel("Breaking Strength",fontsize=15, weight="bold");
plt.ylabel("pdf",fontsize=15, weight="bold");
plt.tick_params(labelsize = 14)
```

```python
#proportion of gunny bags under 3.17
str_val = 3.17
p_br_str = norm.cdf(str_val,mean_breaking_str,std_dev_dist)

#plotting the distribution
plt.figure(figsize=(15,8))
plt.plot(x, br_str_dist_pdf)
# plt.grid()
plt.title("Gunny Bag Breaking Strength Distribution",fontsize=15, weight="bold")

# fill the specified region with red color
idx = np.amax(np.where(x < str_val))
x1 = x[:idx]
plt.axvline(x=str_val, c="r")
plt.text(str_val-0.2,0.15,'less than 3.17',rotation=90,fontsize=15)
plt.fill_between(x1, norm.pdf(x1, mean_breaking_str, std_dev_dist), color="r") 

plt.xlabel("Breaking Strength",fontsize=15, weight="bold");
plt.ylabel("pdf",fontsize=15, weight="bold");
plt.tick_params(labelsize = 14)

print(f"The cumulative probability of breaking strength under {str_val} is {round(p_br_str,4)}")
```

```python
#proportion of gunny bags over 3.6
str_val = 3.6
p_br_str = 1 - norm.cdf(str_val,mean_breaking_str,std_dev_dist)

#plotting the distribution
plt.figure(figsize=(15,8))
plt.plot(x, br_str_dist_pdf)
# plt.grid()
plt.title("Gunny Bag Breaking Strength Distribution",fontsize=15, weight="bold")

# fill the specified region with red color
idx = np.amax(np.where(x <= str_val))
x1 = x[idx:]
plt.axvline(x=str_val, c="r")
plt.text(str_val-0.2,0.17,'More than/ At least 3.6',rotation=90,fontsize=15)
plt.fill_between(x1, norm.pdf(x1, mean_breaking_str, std_dev_dist), color="r") 

plt.xlabel("Breaking Strength",fontsize=15, weight="bold");
plt.ylabel("pdf",fontsize=15, weight="bold");
plt.tick_params(labelsize = 14)

print(f"The cumulative probability of breaking strength over/ at least {str_val} is {round(p_br_str,4)}")
```

```python
#proportion of gunny bags between 5 and 5.5
str_val1 = 5
str_val2 = 5.5
p_br_str1 = norm.cdf(str_val1,mean_breaking_str,std_dev_dist)
p_br_str2 = norm.cdf(str_val2,mean_breaking_str,std_dev_dist)
p_br_str = p_br_str2 - p_br_str1

#plotting the distribution
plt.figure(figsize=(15,8))
plt.plot(x, br_str_dist_pdf)
# plt.grid()
plt.title("Gunny Bag Breaking Strength Distribution",fontsize=15, weight="bold")

# fill the specified region with red color
idx1 = np.amax(np.where(x <= str_val1))
idx2 = np.amax(np.where(x <= str_val2))
x1 = x[idx1:idx2]
plt.axvline(x=str_val2, c="r")
plt.axvline(x=str_val1, c="r")
plt.text(str_val2-0.2,0.05,'Greater than 5 and less than 5.5',rotation=90,fontsize=15)
plt.fill_between(x1, norm.pdf(x1, mean_breaking_str, std_dev_dist), color="r") 

plt.xlabel("Breaking Strength",fontsize=15, weight="bold");
plt.ylabel("pdf",fontsize=15, weight="bold");
plt.tick_params(labelsize = 14)

print(f"The cumulative probability of breaking strength between {str_val1} and {str_val2} is {round(p_br_str,4)}")
```

```python
#proportion of gunny bags NOT between 3 and 7.5
str_val1 = 3
str_val2 = 7.5
p_br_str1 = norm.cdf(str_val1,mean_breaking_str,std_dev_dist)
p_br_str2 = 1-norm.cdf(str_val2,mean_breaking_str,std_dev_dist)
p_br_str = p_br_str2 + p_br_str1

#plotting the distribution
plt.figure(figsize=(15,8))
plt.plot(x, br_str_dist_pdf)
# plt.grid()
plt.title("Gunny Bag Breaking Strength Distribution",fontsize=15, weight="bold")

# fill the specified region with red color
idx1 = np.amax(np.where(x <= str_val1))
idx2 = np.amax(np.where(x <= str_val2))
x1 = x[:idx1]
x2 = x[idx2:]
plt.axvline(x=str_val2, c="r")
plt.axvline(x=str_val1, c="r")
plt.text(4,0.05,'Not Between 3 and 7.5',rotation=0,fontsize=15)
plt.fill_between(x1, norm.pdf(x1, mean_breaking_str, std_dev_dist), color="r") 
plt.fill_between(x2, norm.pdf(x2, mean_breaking_str, std_dev_dist), color="r") 

plt.xlabel("Breaking Strength",fontsize=15, weight="bold");
plt.ylabel("pdf",fontsize=15, weight="bold");
plt.tick_params(labelsize = 14)

print(f"The cumulative probability of breaking strength not between {str_val1} and {str_val2} is {round(p_br_str,4)}")
```

```python
#Declaring known variables
mean_score = 77
score_std = 8.5
```

```python
#Prepare a dataset for visal representation
score_dist = np.random.normal(loc=mean_score, scale=score_std, size=500) 
score_dist_min = score_dist.min()
score_dist_max = score_dist.max()

#x-axis ranges from min to max with 500 steps
x = np.arange(score_dist_min,score_dist_max ,(score_dist_max-score_dist_min)/500 )
score_dist_pdf = norm.pdf(x, mean_score, score_std)

#plotting the distribution
plt.figure(figsize=(15,8))
plt.plot(x, score_dist_pdf)
# plt.grid()
plt.title("Final Examination Grade Distribution",fontsize=15, weight="bold")

plt.xlabel("Grade",fontsize=15, weight="bold");
plt.ylabel("pdf",fontsize=15, weight="bold");
plt.tick_params(labelsize = 14)
```

```python
#probabilty of getting a score below 85
score_val = 85
p_score = norm.cdf(score_val,mean_score,score_std)

#plotting the distribution
plt.figure(figsize=(15,8))
plt.plot(x, score_dist_pdf)
# plt.grid()
plt.title("Final Examination Grade Distribution",fontsize=15, weight="bold")

# fill the specified region with red colo
idx = np.amax(np.where(x < score_val))
x1 = x[:idx]
plt.axvline(x=score_val, c="r")
plt.text(score_val-1,0.037,f'Less than {score_val}',rotation=90,fontsize=15)
plt.fill_between(x1, norm.pdf(x1, mean_score, score_std ), color="r") 

plt.xlabel("Grade",fontsize=15, weight="bold");
plt.ylabel("pdf",fontsize=15, weight="bold");
plt.tick_params(labelsize = 14)


print(f"The probability of getting a score under {score_val} is {round(p_score,4)}")
```

```python
#probabilty of getting a score below 85
score_val1 = 65
score_val2 = 87
p_score1 = norm.cdf(score_val1,mean_score,score_std)
p_score2 = norm.cdf(score_val2,mean_score,score_std)
p_score = p_score2 - p_score1

#plotting the distribution
plt.figure(figsize=(15,8))
plt.plot(x, score_dist_pdf)
# plt.grid()
plt.title("Final Examination Grade Distribution",fontsize=15, weight="bold")

# fill the specified region with red colo
idx1 = np.amax(np.where(x <= score_val1))
idx2 = np.amax(np.where(x <= score_val2))
x1 = x[idx1:idx2]
plt.axvline(x=score_val2, c="r")
plt.axvline(x=score_val1, c="r")
plt.text(70,0.03,f'Score between {score_val1} & {score_val2}',rotation=0,fontsize=15)
plt.fill_between(x1, norm.pdf(x1, mean_score, score_std ), color="r") 

plt.xlabel("Grade",fontsize=15, weight="bold");
plt.ylabel("pdf",fontsize=15, weight="bold");
plt.tick_params(labelsize = 14)


print(f"The probability of getting a score between {score_val1} & {score_val2} is {round(p_score,4)}")
```

```python
#score for 75% to pass
percentile_val = 1-0.75
score_at_percentile = norm.ppf(percentile_val,mean_score,score_std)

#plotting the distribution
plt.figure(figsize=(15,8))
plt.plot(x, score_dist_pdf)
# plt.grid()
plt.title("Final Examination Grade Distribution",fontsize=15, weight="bold")

# Score Line
plt.axvline(x=score_at_percentile, ymax = 0.77 , linestyle = '--', color = 'green')
plt.text(score_at_percentile-0.8,0.01,f'{int(percentile_val*100)} Percentile score ',rotation=90,fontsize=12)
plt.text(score_at_percentile,0,f'{round(score_at_percentile,2)} ',rotation=0,fontsize=15,weight="bold")

plt.xlabel("Grade",fontsize=15, weight="bold");
plt.ylabel("pdf",fontsize=15, weight="bold");
plt.tick_params(labelsize = 14)


print(f"The score for 75 percent to pass is {round(score_at_percentile,4)}")
```
